## Step 4

In this step, we analyze the preferences and behaviors of users in different regions — **North America (NA)**, **Europe (EU)**, and **Japan (JP)** — based on platform, genre, and ESRB rating trends.

```python
# Group sales by platform for each region
platform_sales = df_recent.groupby('platform')[['na_sales', 'eu_sales', 'jp_sales']].sum()

# Get top 5 platforms for each region
top_na = platform_sales['na_sales'].sort_values(ascending=False).head(5)
top_eu = platform_sales['eu_sales'].sort_values(ascending=False).head(5)
top_jp = platform_sales['jp_sales'].sort_values(ascending=False).head(5)

# Combine top platforms
top_platforms = set(top_na.index) | set(top_eu.index) | set(top_jp.index)

# Filter and normalize to get market share
market_share = platform_sales.loc[top_platforms].copy()
market_share['na_share'] = (market_share['na_sales'] / platform_sales['na_sales'].sum() * 100).round(2)
market_share['eu_share'] = (market_share['eu_sales'] / platform_sales['eu_sales'].sum() * 100).round(2)
market_share['jp_share'] = (market_share['jp_sales'] / platform_sales['jp_sales'].sum() * 100).round(2)

# Keep only market share columns
market_share = market_share[['na_share', 'eu_share', 'jp_share']]

# Display the market share table
print(market_share.sort_values(by='na_share', ascending=False))

# Plot market share comparison
market_share.plot(kind='bar', figsize=(12, 6))
plt.title('Top Platforms Market Share by Region')
plt.ylabel('Market Share (%)')
plt.xlabel('Platform')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```

#### Key Observations:

- **PS3** is the most balanced platform globally, holding strong positions in **Europe (25.39%)** and **Japan (19.42%)**, making it a truly international platform.
- In **Japan**, **3DS** is the clear leader with an impressive **32.98%**, far ahead of other platforms, showing a strong regional preference for handheld consoles.
- **Wii** and **PS4** have moderate shares across all regions, indicating consistent global appeal.
- **XOne**, while somewhat popular in NA (**8.24%**), has almost no presence in Japan (**0.11%**), highlighting a regional gap.
- **PC** holds a reasonable share in **Europe (8.18%)** but is virtually absent in Japan.
- **Handheld platforms** such as **PSP** and **PSV** still maintain notable shares in Japan (**13.83%** and **7.16%**, respectively), reflecting the country's ongoing preference for portable gaming.

### Hypotheses Testing

**H₀ (null hypothesis):** The mean user_score is the same for Xbox One and PC.

**H₁ (alternative hypothesis):** The means are different.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the data for Xbox One and PC with non-null user scores
xone_scores = df_recent[(df_recent['platform'] == 'XOne') & (df_recent['user_score'].notna())]['user_score']
pc_scores = df_recent[(df_recent['platform'] == 'PC') & (df_recent['user_score'].notna())]['user_score']

# Create histograms
plt.figure(figsize=(10, 5))
sns.histplot(xone_scores, kde=True, color='skyblue', label='Xbox One', bins=15)
sns.histplot(pc_scores, kde=True, color='orange', label='PC', bins=15)
plt.title('Distribution of User Scores: Xbox One vs PC')
plt.xlabel('User Score')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Describe statiscical information:
print('Xone Score \n',xone_scores.describe())
print()
print('PC Score \n', pc_scores.describe())
```

**"We will use a two-sample independent t-test, assuming unequal variances (Welch’s t-test), with a significance level (α) of 0.05."**

**Key details:**
**"Two-sample independent t-test" → Standard term for comparing means of two unrelated groups.

"Assuming unequal variances" → Clarifies Welch’s adjustment (no assumption of homogeneity of variance).

"Significance level (α) of 0.05" → Conventional phrasing in hypothesis testing.

```python
from scipy.stats import ttest_ind

# Perform Welch’s t-test
t_stat, p_value = ttest_ind(xone_scores, pc_scores, equal_var=False)

# Display results
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Check significance
alpha = 0.05
if p_value < alpha:
    print("We reject the null hypothesis: There is a significant difference in user scores between Xbox One and PC.")
else:
    print("We fail to reject the null hypothesis: No significant difference in user scores between Xbox One and PC.")
```

```python
# Filter data for Action and Sports genres with non-null user scores
action_scores = df_recent[(df_recent['genre'] == 'Action') & (df_recent['user_score'].notna())]['user_score']
sports_scores = df_recent[(df_recent['genre'] == 'Sports') & (df_recent['user_score'].notna())]['user_score']

# Create histograms
plt.figure(figsize=(10, 5))
sns.histplot(action_scores, kde=True, color='blue', label='Action', bins=15)
sns.histplot(sports_scores, kde=True, color='green', label='Sports', bins=15)
plt.title('Distribution of User Scores: Action vs Sports')
plt.xlabel('User Score')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

**H₀ (null hypothesis):** The mean user_score is equal for the Action and Sports genres.

**H₁ (alternative hypothesis):** The mean user_score differs between the genres.

```python
# Perform Welch’s t-test
t_stat2, p_value2 = ttest_ind(action_scores, sports_scores, equal_var=False)

# Display results
print(f"T-statistic: {t_stat2:.3f}")
print(f"P-value: {p_value2:.4f}")

# Check significance
if p_value2 < alpha:
    print("We reject the null hypothesis: There is a significant difference in user scores between Action and Sports genres.")
else:
    print("We fail to reject the null hypothesis: No significant difference in user scores between Action and Sports genres.")
```
